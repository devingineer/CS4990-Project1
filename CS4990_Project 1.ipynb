```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('candy-data.csv')
display(df.head())
```

```python
# Identify candy type columns
candy_type_columns = ['chocolate', 'fruity', 'caramel', 'peanutyalmondy', 'nougat', 'crispedricewafer', 'hard', 'bar', 'pluribus']

# Dictionary to store average popularity for each candy type
popularity_by_candy_type = {}

# Calculate average winpercent for each candy type
for col in candy_type_columns:
    # Filter the DataFrame to include only candies of the current type (where the column value is 1)
    type_df = df[df[col] == 1]
    if not type_df.empty:
        popularity_by_candy_type[col] = type_df['winpercent'].mean()

# Convert the results to a DataFrame for better display
popularity_df = pd.DataFrame(popularity_by_candy_type.items(), columns=['Candy Type', 'Average Winpercent'])
popularity_df = popularity_df.sort_values(by='Average Winpercent', ascending=False).reset_index(drop=True)

display(popularity_df)
```

```python
fig2, ax1 = plt.subplots(figsize=(10, 6))
ax1.bar(popularity_df['Candy Type'], popularity_df['Average Winpercent'], color='blue') # Use specific columns and a basic color
ax1.set_title('Average Candy Popularity by Type') # Corrected title
ax1.set_xlabel('Candy Type') # Corrected x-axis label
ax1.set_ylabel('Average Winpercent')
ax1.set_ylim(0, 100)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Remove all "lines": no grid, no spines, no tick marks
ax1.grid(False)
for spine in ax1.spines.values():
    spine.set_visible(False)
ax1.tick_params(axis="both", length=0)

plt.tight_layout() # Adjust layout to prevent labels from being cut off
plt.show()
```
